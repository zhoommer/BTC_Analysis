```python
from pandas_ods_reader import read_ods
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
path = '/home/mrrobot/Documents/Crypto/BTC_Analysis/BTC_Data.ods'
sheet_idx = 1
data = read_ods(path, sheet_idx)
df = pd.DataFrame(data)
```

```python
df.describe()
```

```python
df
```

```python
x = df[(df['Date'] >= '2018-01-01') & (df['Date'] <= '2019-01-01')]['Low'].mean()
```

```python
y = df[(df['Date'] >= '2019-01-01') & (df['Date'] <= '2020-01-01')]['Low'].mean()
```

```python
z = df[(df['Date'] >= '2020-01-01') & (df['Date'] <= '2021-01-01')]['Low'].mean()
```

```python
sbn.lineplot(
    x = ['2018', '2019', '2020'], 
    y = [x,y,z]
).set_title('Bitcoin Yıllara Göre En Düşük Değer Ortalaması', color='red')
```

```python
sbn.barplot(x = ['2018', '2019', '2020'], y = [x,y,z])
```
